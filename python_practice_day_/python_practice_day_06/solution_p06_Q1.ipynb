```python
steps = [8500, 10200, 5000, 12000, 9500, 15000, 7000]
for w in steps 
    break w >= 10000
#ㄴ 처음에 작성한 코드. :이 빠져있고, if 대신 break을 사용하고있다. 굉장히 멍청한 모습. 또한 걸음 수를 셀 변수도 만들지 않았다. 잠이 덜깬듯
```

```python
steps = [8500, 10200, 5000, 12000, 9500, 15000, 7000]
count = 0
for w in steps:
    if w >= 10000:
        count = count + 1
print(count)
```
